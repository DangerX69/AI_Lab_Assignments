```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
#using pandas to load a .csv file
file_path = "archive/jobs_in_data.csv"

df = pd.read_csv(file_path)
```

```python
#gives tthe number of (rows,column)
df.shape
```

```python
plt.rcParams["figure.figsize"] = (10, 5)
plt.rcParams["xtick.labelsize"] = 7
```

```python
#view the last 10 entries
df.tail(5)
```

```python
df.describe(include="all")
```

```python
# information about the dataset
df.info()
```

```python
# Categorical data
# countplot
sns.countplot(data=df, x='work_year')
plt.title('Count Plot for Number of Employees Grouped by Year')
```

```python
# countplot


sns.countplot(x="job_category", data=df)

plt.xticks(rotation=45)

plt.title('Count Plot for Number of Employees Grouped by their Job Category')
plt.show()
```

```python
 # pie chart

df['company_size'].value_counts().plot(kind='pie', autopct='%0.2f')
plt.legend(loc='upper right', bbox_to_anchor=(1.5, 1), ncol=1)
plt.title('Count of Company size')

plt.show()
```

```python
# Numberical Value

# Histogram treat as categorical data distribution of data
plt.title('Histogram Plot for range count of employees earning salary')
plt.hist(df['salary_in_usd'])
```

```python
# Boxplot
# 5 number summary
# median , percentile, min=Q1-1.5*IQR, max=Q3+1.5*IQR, IQR = (Q1-Q3), outlier
plt.title('Box Plot for Salary')
sns.boxplot(df['salary_in_usd'])
```

```python
df.describe()
```

```python
# Bar plot (Numerical - Categorical)

sns.barplot(x='employment_type',y='salary',hue='experience_level',data=df)
plt.title("Salaries Based on employement type and experience level")
```

```python
sns.barplot(x='work_setting',y='salary',hue='job_category',data=df)
plt.legend(loc='upper right', bbox_to_anchor=(1.37, 1), ncol=1)
plt.title("Salaries Based on work setting and job category")
```

```python
#line graph to show increaing salaries with increase in year
sns.lineplot(x='work_year',y='salary',data=df)
plt.title("Line Plot for Salary vs Year")
```

```python
sns.heatmap(df.pivot_table(values='salary',index='work_setting',columns='work_year'))
plt.title("Heat-Map Showing Salaries based on year and type of work setting")
```

```python
sns.heatmap(df.pivot_table(values='salary',index='company_size',columns='job_category'))
plt.title("Heat-Map Showing Salaries based on category and company size")
```

```python
sns.scatterplot(x='company_location', y='salary_in_usd',hue='job_category', style='work_year',size='experience_level' ,data=df)
plt.legend(loc='upper right', bbox_to_anchor=(1.37, 1), ncol=1)
plt.xticks(rotation=85)
plt.title("Scatter Plot Showing Salaries based on Company Location, along with Category, Experience and Year")
plt.show()
```
